$\textbf{Directed Graph}$

A directed graph, also known as a digraph, is a graph where edges have a direction associated with them. In this example, nodes 1, 2, and 3 are connected by directed edges, indicating the direction of the relationship between nodes. For example, node 1 has an outgoing edge to node 2, and node 2 has an outgoing edge to node 3.

```python
import networkx as nx
import matplotlib.pyplot as plt

simple_graph = nx.Graph()
simple_graph.add_edges_from([(1, 2), (2, 3), (3, 1)])
nx.draw(simple_graph, with_labels=True, node_color='lightgreen', node_size=800)
plt.title("Simple Graph")
plt.show()
```

$\textbf{Simple Graph}$

A simple graph is an undirected graph that does not contain any self-loops (edges that connect a node to itself) or multiple edges between the same pair of nodes. In this example, nodes 1, 2, and 3 are connected by undirected edges, forming a simple graph.

```python
import networkx as nx
import matplotlib.pyplot as plt

complete_graph = nx.complete_graph(5)
nx.draw(complete_graph, with_labels=True, node_color='lightcoral', node_size=800)
plt.title("Complete Graph")
plt.show()
```

$\textbf{Complete Graph}$

A complete graph is a graph in which every pair of distinct vertices is connected by a unique edge. In this example, a complete graph with 5 nodes is generated, resulting in each node being connected to every other node by an edge, forming a fully connected graph.

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
G.add_nodes_from([1, 2, 3], bipartite=0) 
G.add_nodes_from(['A', 'B', 'C'], bipartite=1)
G.add_edges_from([(1, 'A'), (1, 'B'), (2, 'B'), (2, 'C'), (3, 'A')])

pos = nx.bipartite_layout(G, nodes=[1, 2, 3])  
nx.draw(G, pos, with_labels=True)

plt.show()
```

$\textbf{Bipartite Graph}$

A bipartite graph is a type of graph where the set of nodes can be divided into two disjoint sets such that no two nodes within the same set are connected by an edge. . In this example, nodes are divided into two distinct sets: set 1 (nodes 1, 2, and 3) and set 2 (nodes 'A', 'B', and 'C'), with edges only connecting nodes from different sets, visually represented with light blue and light green nodes, showcasing relationships between entities of different types.

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5])
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5), (5, 1)])


sub_nodes = [1, 2, 3]
subgraph = G.subgraph(sub_nodes)


plt.figure(figsize=(10, 5))

plt.subplot(121)
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray')
plt.title('Original Graph')


plt.subplot(122)
nx.draw(subgraph, with_labels=True, node_color='lightgreen', edge_color='black')
plt.title('Subgraph (Subset of Nodes 1, 2, 3)')

plt.tight_layout()
plt.show()
```

$\textbf{Subgraphs}$



A subgraph is a subset of the nodes and edges of a larger graph that forms a graph structure on its own. In this example the original graph is a simple undirected graph with nodes 1, 2, 3, 4, and 5 connected by edges, while the subgraph comprises nodes 1, 2, and 3 along with the edges connecting them, forming a connected graph structure within the selected subset.

```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.Graph()
graph.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5)])

degree = dict(graph.degree())
print("Degree of each vertex:", degree)

nx.draw(graph, with_labels=True, node_color='lightgrey', node_size=800)
plt.title("Graph with Degrees")
plt.show()
```

$\textbf{Degree of a Vertex}$

$\textbf{Walk}$



a walk is a sequence of vertices and edges that starts at a vertex, follows edges from one vertex to another, and ends at a vertex. In the given graph, an example of a Walk could be $A \to B \to D \to E \to F$, meaning we start at node A, move to node B, then to node D, and so on, following the edges of the graph until we reach node F.

```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.Graph()
graph.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1)])

trail_nodes = [1, 2, 3, 4, 5, 1]  # Repeating the starting node to form a closed trail
trail = nx.Graph(graph.subgraph(trail_nodes))
nx.draw(trail, with_labels=True, node_color='lightgreen', node_size=800)
plt.title("Trail")
plt.show()
```

$\textbf{Trail}$

A trail in a graph is a walk in which no edge is repeated. In this example, a trail is formed by selecting a sequence of nodes (1, 2, 3, 4, 5, 1) from the original graph such that no edge is repeated, representing a path that visits each edge exactly once.

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
nodes = [1, 2, 3, 4, 5, 6, 7, 8]
edges = [(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (3, 7), (4, 8), (5, 8), (6, 8), (7, 8)]
G.add_nodes_from(nodes)
G.add_edges_from(edges)


nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Path}$

A "path" in graph theory is a trail in which all vertices are distinct. An example of a path in the example below could be: $1 \to 3 \to 6 \to 8 \to 5$:

```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.Graph()
graph.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1)])

closed_walk_nodes = [1, 2, 3, 4, 5, 1]  # Repeating the starting node to form a closed walk
closed_walk = nx.Graph(graph.subgraph(closed_walk_nodes))
nx.draw(closed_walk, with_labels=True, node_color='lightblue', node_size=800)
plt.title("Closed Walk")
plt.show()
```

$\textbf{Closed Walk}$

A closed walk in a graph is a walk that starts and ends at the same node. In this example, a closed walk is formed by selecting a sequence of nodes (1, 2, 3, 4, 5, 1) from the original graph, where the last node is the same as the first one, forming a loop.

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
nodes = ['X', 'Y', 'Z', 'W']
edges = [('X', 'Y'), ('Y', 'Z'), ('Z', 'W'), ('W', 'X'), ('X', 'Z')]

G.add_nodes_from(nodes)
G.add_edges_from(edges)


nx.draw(G, with_labels=True)
plt.show()

```

$\textbf{Circuit}$

A circuit is a closed walk in a graph where all edges are distinct except for the first and last edge, which are the same. In the graph below the cycle $X \to Y \to Z \to W \to X$ forms a circuit

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
nodes = ['P', 'Q', 'R', 'S', 'T']
edges = [('P', 'Q'), ('Q', 'R'), ('R', 'S'), ('S', 'T'), ('T', 'P'), ('P', 'S'), ('Q', 'T')]

G.add_nodes_from(nodes)
G.add_edges_from(edges)

nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Simple Circuit}$

A simple circuit is a closed walk in a graph where all edges and vertices are distinct, except for the first and last vertex, which are the same. In the graph below, the cycle $P \to Q \to T \to S \to P$ forms a simple circuit.

```python
import networkx as nx
import matplotlib.pyplot as plt

connected_graph = nx.Graph()
connected_graph.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1)])

nx.draw(connected_graph, with_labels=True, node_color='lightcyan', node_size=800)
plt.title("Connected Graph")
plt.show()
```

$\textbf{Connected Graph}$

$\textbf{Euler's Circuit}$

An Euler's circuit is a circuit that uses every edge exactly once and returns to the starting node. In this example, a complete graph with 5 nodes is created, which has an Euler's circuit since every node is connected to every other node forming a circuit that traverses each edge exactly once and returns to the starting node.

```python
import networkx as nx
import matplotlib.pyplot as plt

euler_graph = nx.Graph()
euler_graph.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 3)])

nx.draw(euler_graph, with_labels=True, node_color='lightpink', node_size=800)
plt.title("Euler's Trail")
plt.show()
```

$\textbf{Euler's Trail}$

An Euler's trail is a trail that uses every edge exactly once. Unlike an Euler's circuit, it does not necessarily return to the starting node. In this example, a graph is created with an Eulerian trail by adding an extra edge between nodes 1 and 3 to break the circuit. This trail traverses each edge exactly once but does not return to the starting node.



```python
import networkx as nx
import matplotlib.pyplot as plt

hamiltonian_graph = nx.Graph()
hamiltonian_graph.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 3), (2, 4), (3, 5)])

nx.draw(hamiltonian_graph, with_labels=True, node_color='lightblue', node_size=800)
plt.title("Hamiltonian Circuit")
plt.show()
```

$\textbf{Hamiltonian Circuit}$

A Hamiltonian circuit is a circuit that visits every vertex exactly once and returns to the starting vertex. In this example, a graph is created with a Hamiltonian circuit by adding additional edges to ensure that every node is connected to every other node forming a circuit that visits each node exactly once and returns to the starting node.
